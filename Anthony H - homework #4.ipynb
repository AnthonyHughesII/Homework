```python
#1.Create an if statement: if 'age' is greater than or equal to 25, 
#print "Renting a car is more affordable", however if 'age' is less 
# than 25, print "Renting a car is very expensive."
import random

age = random.randint(16, 125)
print("You are", age, "years old.")
if (age >= 25):
    print("Renting a car is more affordable.")
else:
    print("Renting a car is very expensive.")
```

```python
#2.Create and chain an if-else statement: if 'age' is greater than or equal to 25, 
#print "Renting a car is more affordable." If 'age' is less than 25 
#but greater than or equal to 18, print "Renting a car is very expensive." 
#Finally, if age is less than 18, print "You cannot legally rent a car." 
import random

age = random.randint(16, 125)
print("You are", age, "years old.")
if (age >= 25):
    print("Renting a car is more affordable.")
elif (age >= 18):
    print("Renting a car is very expensive.")
else:
    print("You cannot legally rent a car")
```

```python
#3.Loop over the following string to (1) count all the characters 
#in the string and (2) print out all the vowels -- 
#"The quick brown fox jumps over the lazy dog"
statement = "The quick brown fox jumps over the lazy dog."
count = 0
vowels = "aeiouAEIOU"
for char in statement:
    if char in vowels:
        print(char)
    count += 1

print("The string has", count, "characters.")
```

```python
#4.Write a nested loop that prints out every piece of clothing from the couture list, 
#in every fashionable color from the panettone set: 
#couture = ["trousers", "blouse", "bandana", "cumber band", "blazer", "vest", "french beret", "scarf", "stole"] 
#and panettone = {"cerise", "fuchsia", "aqua", "maple", "auburn", "burnt sienna", "gunmetal blue", "Dark Sapphire"}

couture = {"trousers", "blouse", "bandana", "cumber band", "blazer", "vest", "french baret", "scarf", "stole"}
panettone = {"cerise", "fuchsia", "aqua", "maple", "auburn", "burnt sienna", "gunmetal blue", "Dark Sapphire"}

for item in couture:
    for color in panettone:
        print(item,"is available in", color)
```
