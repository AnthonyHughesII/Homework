```python
#10-12.
import pandas as pd
import numpy as np

array1 = np.array([[25, 26]])
array2 = np.array([[11, 2], [13, 4]])
array3 = np.array([[7, 81], [5, 6], [11, 12]])

newArray = np.concatenate([array1, array2, array3], axis = 0)
print("New Array:\n",newArray)

newArray.sort()
sortedArray = newArray
print("Sorted Array:\n", sortedArray)
reshapedArray = sortedArray

indexZero = reshapedArray[0]
indexOne = reshapedArray[1]
indexTwo = reshapedArray[2]
indexThree = reshapedArray[3]

print("First Element:", indexZero)
print("Second Element:", indexOne)
print("Third Element:", indexTwo)
print("Fourth Element:", indexThree)
```
